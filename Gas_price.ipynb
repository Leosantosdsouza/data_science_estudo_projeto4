```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Csv/gas_prices.csv')
gas
```

```python
plt.title('Gas que foi gasto esses dias ai tudo em dinheiro')

plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], 'y.-', label = 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'USA')
plt.plot(gas['Year'], gas['Japan'], 'g.-', label = 'Japan')

plt.xlabel('Years')
plt.ylabel('Dinheiros')

plt.xticks(gas['Year'][::2])

plt.legend()
plt.show()
```

```python
plt.title('Gas que foi gasto esses dias ai tudo em dinheiro')

list_country = ['Australia','USA','Italy', 'Mexico', 'France', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xlabel('Years')
plt.ylabel('Dinheiros')

plt.xticks(gas['Year'][::2])

plt.legend()
plt.savefig('country.png')
plt.savefig('country.PDF')
plt.show()
```
